```python
ratings.isnull().sum(axis = 0)
```

```python
 print("max movie ID", ratings.movieId.max())
```

```python
rating_pivot = ratings.pivot_table(index='userId',columns='movieId',values='rating',)
rating_pivot
```

```python
## sparse matrix

rating_pivot.fillna(value = 0 , inplace=True)
rating_pivot
```

### Answer 3

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
distances = pairwise_distances(rating_pivot,metric='cosine')
distances
```

### Answer 7


```python
movies = pd.read_csv('movies.csv')
movies
```

## Data Cleaning

```python
movies.drop('genres',axis=1,inplace=True)
```

```python
# Separting movie title and year part using split function.
split_values = movies['title'].str.split("(", n = 1, expand = True)
# setting 'movie_title' values to title part.
movies.title = split_values[0]
# creating 'release_year' column.
movies['release_year'] = split_values[1]
# Cleaning the release_year series.
movies['release_year'] = movies.release_year.str.replace(')','')
```

```python
movies
```

```python
merge = pd.merge(ratings,movies,on='movieId')
merge
```

```python
merge[(merge.userId==2) | (merge.userId==338)].movieId
```

```python
import numpy as np
c1 = merge['title'][merge.userId==2]
c2 = merge['title'][merge.userId==338]
common = np.intersect1d(c1,c2)
common
```

These are the movies that users 2 and 338 have watched

```python
cond=((merge.title=='Kill Bill: Vol. 1 ') | (merge.title=='Shawshank Redemption, The ')) & ((merge.userId==2) | (merge.userId==338))
Q7 = merge[cond]
Q7
```

```python
Q7 = Q7.pivot(index='title',columns='userId',values='rating')
Q7
```

This is how the users 2 and 338 have rated the movies they've watched together
